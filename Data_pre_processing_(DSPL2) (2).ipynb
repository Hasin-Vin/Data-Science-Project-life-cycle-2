```python
admin4_df = admin4_df.drop(columns=[col for col in admin4_df.columns if '_SI' in col or '_TA' in col])

# Step 3: Clean column names by removing '_EN' suffix
admin4_df.columns = [col.replace('_EN', '') for col in admin4_df.columns]

# Step 4: Check column names (optional but useful for debugging)
print("Columns after renaming:", admin4_df.columns.tolist())

# Step 5: Drop rows where Admin4_Name or Admin4_Code is missing
admin4_df.dropna(subset=["admin4Name_en", "admin4Pcode"], inplace=True)

# Step 6: Display the first few rows
admin4_df.head()
```

```python
# Step 4: Standardize text fields
for col in admin4_df.select_dtypes(include="object"):
    admin4_df[col] = admin4_df[col].astype(str).str.strip().str.title()
```

```python
# Step 5: Convert Record_Date to datetime
admin4_df["date"] = pd.to_datetime(admin4_df["date"], errors='coerce')
```

```python
# Step 6: Save cleaned data
output_path = "cleaned data FOR EDA ADMIN BOARDERS.csv"
admin4_df.to_csv(output_path, index=False)
print(f"✅ File saved at: {output_path}")
```
